## **Read the image in RGB format**

```python
image = cv2.imread(image_path)
image1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image(data=cv2.imencode('.jpg', image1)[1].tobytes()))
```

## **Convert the image to grayscale**

```python
grayscale_image = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
display(Image(data=cv2.imencode('.jpg', grayscale_image)[1].tobytes()))
```

## **Invert the grayscale image**

```python
inverted_image = cv2.bitwise_not(grayscale_image)
display(Image(data=cv2.imencode('.jpg', inverted_image)[1].tobytes()))
```

## **Apply Gaussian blur to the inverted image**

```python
blurred_image = cv2.GaussianBlur(inverted_image, (27, 27), 0)
display(Image(data=cv2.imencode('.jpg', blurred_image)[1].tobytes()))
```

## **Sketch of image**
-  Create the pencil sketch by dividing the grayscale image by the inverted blurred image

```python
pencil_sketch = cv2.divide(grayscale_image, blurred_image, scale=200.0)
display(Image(data=cv2.imencode('.jpg', pencil_sketch)[1].tobytes()))
```
